# **6. Data Pre-Processing**
---

- Here we will perform ** Data Processing** on our dataset to make data usable for EDA.

- We are utilizing dataset on which pandas profiling is already applied once.

- Let's correct the dataset first.

<a name = Section6></a>
- As first step we should be renaming the columns Runtime (Minutes) and Revenue (Millions) as any further operations would be difficult to follow. These names should be made one word names. 
- Secondly, all the column names should be made lowecase as that is the convention followed most of the times.

```python
data.rename(columns={'Runtime (Minutes)': 'Durinmins','Revenue (Millions)': 'Revinmil'})
```

```python
data.rename(columns={'Runtime (Minutes)': 'Durinmins','Revenue (Millions)': 'Revinmil'},inplace=True)
```

```python
data.columns = map(str.lower,data.columns)
```

```python
data.columns
```

```python
data[data['revinmil'].isnull()]
```

- revinmil has 128 missing values
    - We will replace missing revinmil values by taking **median** of every **'genre'**.
    - Let's use **transform()** method for finding the **meadian** values. 
    - In some cases **'genre'** will be unique and will have missing values. Those values will be filled by the median of all the non null values to be safe.

```python
data['revinmil'] = data.groupby('genre')['revinmil'].transform(lambda x : x.fillna(x.median()))
```

```python
data[data['revinmil'].isnull()]
```

```python
(data[data['revinmil'].notnull()]).median()['revinmil']
```

```python
data['revinmil'] = data['revinmil'].fillna(43.135)
```

```python
data[data['revinmil'] == 43.135]
```

- metascore has missing values
    - metascore is given to the movies that are reviewed by world's best critics. 
    - We have 68 movies where the metascore is missing. This is a small number compared the size of the dataset. And so rather than dropping those records we have decided to replace the metascore. 
    - We will replace missing metascore values by taking **median** of every **'genre'**. This is exactly like what we did for replacing revinmil missing values. 
    - Let's use **transform()** method for finding the **meadian** values. 
    - Despite of this step in case if there are still any missing values in metascore, then we will have to manually replace them. 

```python
data[data['metascore'].isnull()]
```

```python
data['metascore'] = data.groupby('genre')['metascore'].transform(lambda x : x.fillna(x.median()))
```

```python
(data[data['metascore'].notnull()]).median()['metascore']
```

```python
data['metascore'] = data['metascore'].fillna(60.0)
```

```python
data[data['metascore'].isnull()]
```

```python
data['year'] = pd.to_datetime(data['year'],format='%Y')
```

```python
data['yr'] = data['year'].dt.year
```

- year appears to be an integer column. 
    - We will change the datatype of **year** to **datetime**
    - also created an **integer column** only having **year (yr)** so as to use the same for plotting purpose.

```python
data.info()
```

- Finally all the **float** numbers will be converted to **two decimals** for ease of operations with the help of **round()** method.  

```python
data = data.round(decimals=2)
```

# **8. Exploratory Data Analysis**
---


**Exploratory Data Analysis(EDA)** is an **approach to analyzing data sets** to **summarize** their **main characteristics**, often with **visual methods**. 


- It includes **cleaning, munging, combining, reshaping, slicing, dicing, and transforming data** for **analysis** purpose.


- The **primary goal of EDA** is to **maximize the analyst's insight into a data set** and into the **underlying structure of a data set**, while **providing all of the specific items** that an **analyst would want to extract from a data set**, such as:
 <br><br> 
  - A list of **outliers**.<br><br>
  
  - A **good-fitting model**.  <br><br>
  
  - **Estimates for parameters**.<br><br>
  
  - **Uncertainties** for those **estimates**.<br><br>
  
  - A **ranked list** of **important factors**.<br><br>
  
  - **Conclusions** as to whether **individual factors are statistically significant**.<br><br>
  
  - A **sense of robustness of conclusions**.<br><br>
  
  - **Optimal settings**.<br><br>
